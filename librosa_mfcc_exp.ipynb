```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
import scipy.io.wavfile as w

#sr,x=w.read('o_o3.wav')
x,sr=(librosa.core.load('GN00000439_0043426.wav',sr=None))
print x*10,sr
mfccs = librosa.feature.mfcc(y=x, sr=sr, n_fft=2048, n_mfcc=50, fmax=sr/2, hop_length=sr/100)
#mfccs = (mfccs - np.mean(mfccs,0))/(np.std(mfccs,0) + 1e-6)
#mfccs = (mfccs - np.mean(mfccs))/(np.std(mfccs) + 1e-6)
mfccs = (mfccs - np.mean(mfccs))/(np.std(mfccs) + 1e-6)

mfccs = mfccs[3:,:]

print mfccs.shape, type(mfccs)
#plt.plot(x*10)
plt.pcolormesh(mfccs, cmap='hot')
plt.show()

#print mfccs[:,0].size
print mfccs[:,0]
```

```python
x,sr=(librosa.core.load('GN00000439_0043426_1.wav',sr=None))
print x,sr
mfccs = librosa.feature.mfcc(y=x, sr=sr, n_fft=2048, n_mfcc=50, fmax=sr/2, hop_length=sr/100)
mfccs = (mfccs - np.mean(mfccs,0))/(np.std(mfccs,axis=0) + 1e-6)
print mfccs.shape
plt.pcolormesh(mfccs, cmap='hot')
plt.show()
```

```python
S = librosa.feature.melspectrogram(y=x, sr=sr, n_mels=50, n_fft=2048,  hop_length=sr/100)
S=librosa.power_to_db(S,ref=np.max)
S = (S-np.mean(S))/(np.std(S) + 1e-6)
print S.shape
plt.pcolormesh(S)
plt.show()
```
